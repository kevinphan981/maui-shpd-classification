# General Data Collection
From 2012 to 2024, the data vars look consistent throughout with exception of the "thing0", which I think just counts the index. We can remove that and select the columns we want.

### Plan
1. Maybe make a new "consolidated" sheet in the workbook containing everything but the beans(?) and thing0 columns. We would have to fetch everything but that is fine. 
2. Clean the NULL sf that can show up (supposed to be Flr Area), then also move the entries in the "Units" var that contain "sf" (square feet) into the Flr Area var.
3. Force everything to be lowercase while you're at it.

```python
year = "2024"
path = "../raw-data/my_copy_building-permits.xlsx"
df = pd.read_excel(path, sheet_name = year)

#make all columns lowercase
df.columns = df.columns.str.lower()

#make everything lowercase
df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

#must drop the thing0 column since it is redundant. also drops anything with "beans"
# df.drop(df.columns[1], axis = 1, inplace= True)
df.drop(list(df.filter(regex = 'beans|thi')), axis = 1, inplace= True)

```
